#### Feature Engineering - estandarización, eliminación, creación de nuevas columnas

```python
df2 = df.copy()
```

```python
# Eliminamos las columnas como se determinó en el EDA
df2 = df2.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'], axis=1)
```

```python
# Procesamos los valores nulos como se determinó en el EDA
df2['NumCompaniesWorked'] = df2['NumCompaniesWorked'].fillna(0)

wya = df2.groupby(by='NumCompaniesWorked')['TotalWorkingYears'].mean().round(decimals=0).reset_index()

df2['TotalWorkingYears'] = np.where(df2['TotalWorkingYears'].isna(), df2['NumCompaniesWorked'].map(
    wya.set_index('NumCompaniesWorked')['TotalWorkingYears'],na_action='ignore'),df2['TotalWorkingYears'])
```

```python
# Cambiamos las columnas con data type "object" a "category"
df2['Attrition'] = df2['Attrition'].astype('category')

df2['BusinessTravel'] = df2['BusinessTravel'].astype('category')
df2['Department'] = df2['Department'].astype('category')

df2['EducationField'] = df2['EducationField'].astype('category')
df2['Gender'] = df2['Gender'].astype('category')

df2['JobRole'] = df2['JobRole'].astype('category')
df2['MaritalStatus'] = df2['MaritalStatus'].astype('category')
```

```python
# Las columnas con datatype float64 no es necesario que muestren valores decimales
df2['NumCompaniesWorked'] = df2['NumCompaniesWorked'].astype('int64')
df2['TotalWorkingYears'] = df2['TotalWorkingYears'].astype('int64')
```

```python
# Generamos diccionarios para poder interpretar los resultados más adelante
# Aqui mismo ya modificamos los dtypes "object" a "category"
attr_d = dict(enumerate(df2['Attrition'].astype('category').cat.categories))
bstr_d = dict(enumerate(df2['BusinessTravel'].astype('category').cat.categories))
dept_d = dict(enumerate(df2['Department'].astype('category').cat.categories))
edfd_d = dict(enumerate(df2['EducationField'].astype('category').cat.categories))
gend_d = dict(enumerate(df2['Gender'].astype('category').cat.categories))
jbrl_d = dict(enumerate(df2['JobRole'].astype('category').cat.categories))
mast_d = dict(enumerate(df2['MaritalStatus'].astype('category').cat.categories))
```

```python
# Utilizamos cat codes para asignar números a las columnas categoricas
df2['Attrition'] = df2['Attrition'].cat.codes
df2['BusinessTravel'] = df2['BusinessTravel'].cat.codes
df2['Department'] = df2['Department'].cat.codes
df2['EducationField'] = df2['EducationField'].cat.codes
df2['Gender'] = df2['Gender'].cat.codes
df2['JobRole'] = df2['JobRole'].cat.codes
df2['MaritalStatus'] = df2['MaritalStatus'].cat.codes
```

```python
plt.figure(figsize=(22,8))

mask = np.triu(np.ones_like(df2.select_dtypes(include=['int8','int64']).corr(), dtype=bool))
heatmap = sns.heatmap(df2.select_dtypes(include=['int8','int64']).corr(), mask=mask, annot=True, cmap='YlGnBu')
```

Las variables relacionadas con años tienen alta correlación entre ellas. Sin embargo, no considero necesario eliminarlas ya que todas ellas son factores determinantes en el grado de attrition de un empleado.

```python
# Estandarizamos los valores de la columna "MonthyIncome" ya que su magnitud es muy distinta de las otras columnas
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df2['MonthlyIncome'] = scaler.fit_transform(df2[['MonthlyIncome']])
```

#### Aplicación de modelo de regresión logística a la variable Attrition

```python
from sklearn.model_selection import train_test_split

X = df2.drop('Attrition', axis=1)
y = df2['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
# Como se observó antes, la variable que buscamos predecir con la regresión está desbalanceada en si
df2['Attrition'].value_counts()
```

```python
# Para compensar el desbalanceo, utilizaremos class_weight='balanced' en la regresión logística
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', class_weight='balanced')
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
```

#### Visualización de los resultados del modelo de clasificación

```python
z = X_test.copy()
z['Attrition obsv'] = y_test
z['Attrition pred'] = y_predict
z[['Attrition obsv', 'Attrition pred']].sample(10)
```

#### Matriz de Confusión y Métricas del modelo

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_test, y_predict, labels=[1,0])
df_cm = pd.DataFrame(cm, index=[i for i in ['Obsv=0','Obsv=1']],
                     columns= [i for i in ['Pred=0','Pred=1']])

plt.figure(figsize=(7,4))
sns.heatmap(df_cm, annot=True, cmap='Accent', fmt='d').set_title('Confusion Matrix')
```

```python
model_score = metrics.accuracy_score(y_test, y_predict)
precision_score = metrics.precision_score(y_test, y_predict)
recall_score = metrics.recall_score(y_test, y_predict)
f1_score = metrics.f1_score(y_test, y_predict)
roc_auc_score = metrics.roc_auc_score(y_test, y_predict)

print('Metricas del Modelo')
print('----------------------------------------------')
print('Accuracy = {0:0.2f}'.format(model_score))
print('Precision = {0:0.2f}'.format(precision_score))
print('Recall Score = {0:0.2f}'.format(recall_score))
print('F1 Score = {0:0.2f}'.format(f1_score))
print('ROC AUC Score = {0:0.2f}'.format(roc_auc_score))
```
